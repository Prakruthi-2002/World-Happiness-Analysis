```python
import pandas as pd

df = pd.read_csv('2019.csv')

df.head()
```

```python
print("Shape of dataset:", df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
print("Top 10 Happiest Countries")

df[['Country or region','Score']].sort_values(
    by='Score',
    ascending=False
).head(10)
```

## Visualization 1: Top 10 Happiest Countries

```python
import matplotlib.pyplot as plt

top10 = df.sort_values(by='Score', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10['Country or region'], top10['Score'])
plt.title('Top 10 Happiest Countries')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

## Visualization 2: Distribution of Happiness Scores

```python
plt.figure(figsize=(8,5))
plt.hist(df['Score'], bins=10)
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

## Visualization 3: GDP per Capita vs Happiness Score

```python
plt.figure(figsize=(8,5))
plt.scatter(df['GDP per capita'], df['Score'])
plt.title('GDP per Capita vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()
```

## Visualization 4: Top 5 Countries by Happiness Score

```python
top5 = df.sort_values(by='Score', ascending=False).head(5)

plt.figure(figsize=(8,8))
plt.pie(
    top5['Score'],
    labels=top5['Country or region'],
    autopct='%1.1f%%'
)
plt.title('Top 5 Happiest Countries')
plt.show()
```

## Visualization 5: Correlation Heatmap

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')
plt.show()
```

## Visualization 6: Top 10 Countries Happiness Trend

```python
top10 = df.sort_values(by='Score', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.plot(
    top10['Country or region'],
    top10['Score'],
    marker='o'
)

plt.title('Top 10 Happiness Scores')
plt.xlabel('Country')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

# Insights Report


### Insight 1

Finland achieved the highest happiness score (7.769), followed by Denmark and Norway. The majority of the top 10 happiest countries are European nations, indicating that strong economic conditions, healthcare systems, and social welfare policies contribute significantly to overall happiness.

### Insight 2

South Sudan recorded the lowest happiness score (2.853), while Central African Republic and Afghanistan were also among the least happy countries. This suggests that economic instability, conflict, and limited access to essential services have a major negative impact on people's well-being.

### Insight 3

The correlation between GDP per capita and Happiness Score is 0.7939, which indicates a strong positive relationship. Countries with higher economic prosperity generally tend to report higher happiness levels, as shown in both the correlation analysis and scatter plot.

### Insight 4

Social Support has a strong positive correlation (0.78) with Happiness Score. Countries such as Iceland, Finland, Norway, and Denmark rank highly in both happiness and social support, highlighting the importance of strong family, community, and social networks.

### Insight 5

Healthy Life Expectancy shows one of the strongest relationships with Happiness Score (correlation = 0.78). The heatmap reveals that countries with better health conditions and longer life expectancy generally achieve higher happiness scores, emphasizing the importance of healthcare and quality of life.


Most Surprising Finding

The most surprising finding was that some countries with the highest GDP per capita, such as Qatar and Kuwait, were not among the top 10 happiest countries. This indicates that economic wealth alone does not guarantee happiness. Factors such as social support, health, freedom of choice, and overall quality of life also play a critical role in determining a country's happiness level.
